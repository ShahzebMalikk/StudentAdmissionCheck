# Importing all Important Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

# Importing dataset and checking

```python
df = pd.read_csv('grad_add.csv')
```

```python
print(df.shape)
```

```python
df.info()
```

```python
print(df.head(30))
```

```python
print(df.tail(30))
```

# Cleaning up the Columns

removing the serial number coloumn because pandas already provide us and making all Indexes shorter

```python
df = df.drop(['Serial No.'], axis=1 )
```

```python
new_names = {'GRE Score':'gre', 'TOEFL Score':'toefl', 'University Rating':'rating', 'SOP':'sop', 'LOR':'lor', 'CGPA':'gpa', 'Research':'research', 'Chance of Admit':'chance'}
df.rename(columns=new_names, inplace=True)
df.head(10)
```

# Sub Plot

```python
fig, ax = plt.subplots(figsize=(8,6))
sns.countplot(df['rating'])
plt.title('University Rating')
plt.ylabel('Number of Applicants')
```

# Box Plot

using df.describe() to see the high-level distribution of data

```python
df.describe()
```

```python
plt.figure(1, figsize=(14,4))
for i in range(1,6):
    plt.subplot(1,5,i)
    plt.boxplot(df[df.columns[i]])
    plt.title(df.columns[i])
plt.show()
```

# Histogram Plot

Plot histograms for main columns to show detailed distribution of data

```python
df[['gre','toefl','rating','gpa','research','chance']].hist(figsize=(14, 9),bins=16,linewidth='1',edgecolor='k',grid=False)
plt.show()
```

# Histogram Correlation Plot

Calculating the correlation between each data column and Chance of Acceptance.

it will help us to find that which thing is most important to score better to get the acceptance.

```python
correlation = df.corr()['chance']
plt.bar(df.columns, correlation, color='blue')
plt.show()
print(correlation)
```

# Scatter Plot

 Plot Relations between GRE & TOEFL

```python
df.plot(kind='scatter', x='gre', y='toefl', color='Blue',figsize=(10, 6))
plt.xlabel("GRE")
plt.ylabel("TOEFL")
plt.show()
```

Plot Relations between GPA & GRE

```python
df.plot(kind='scatter', x='gpa', y='gre', color='green',figsize=(10, 6))
plt.xlabel("GPA")
plt.ylabel("GRE")
plt.show()
```

# Plot relations between Chance and other Features

```python
for i in range(1,7):
    df.plot(kind='scatter', x='chance', y=df.columns[i]) 
    plt.xlabel("Chance")
    plt.ylabel(df.columns[i])
    plt.show()
```

# Plotting Heatmap for checking the chances

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(df.corr(), annot=True, cmap='Blues')
```
